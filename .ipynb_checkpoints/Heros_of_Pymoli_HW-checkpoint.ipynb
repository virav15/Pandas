```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import json as js
```

```python
# File to Load (Remember to Change These)
file_to_load = "raw_data/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
##
purchase_data.head()
```

```python
# PLAYER COUNT

players = {
    'Total number of Players': [purchase_data['SN'].nunique()]
}

players_count = pd.DataFrame(players)
players_count
###
```

```python
# PURCHASING ANALYSIS (TOTAL)

# Number of Unique Items
total_unique_items = purchase_data['Item Name'].nunique()

# Average Purchase Price
total_avg_price = purchase_data['Price'].mean()

#Total Number of Purchases
total_purchases = len(purchase_data.index)

#Total Revenue
total_revenue = purchase_data['Price'].sum()

#Analysis
purchase_analysis_df = pd.DataFrame.from_dict({"Number of unique items":[total_unique_items],"Average Price":[total_avg_price],"Number of Purchases":[total_purchases],"Total Revenue":[total_revenue]}, orient='columns')
purchase_analysis_df["Total Revenue"] = purchase_analysis_df["Total Revenue"].map("$ {:,.2f}".format)
purchase_analysis_df["Average Price"] = purchase_analysis_df["Average Price"].map("$ {:,.2f}".format)

purchase_analysis_df
```

```python
#Gender Demographics

# Total Count
total_count = len(purchase_data["SN"].unique())

#Percentage and Count of Male Players
#Male = data1.loc[data1['Gender'] =='Male',:]
Male1= purchase_data.groupby(['Gender']).get_group(('Male'))
Male2= len(Male1['SN'].unique())
MalePercent= round((Male2/total_count)*100,2)

#Percentage and Count of Female Players
Female1= purchase_data.groupby(['Gender']).get_group(('Female'))
Female2= len(Female1['SN'].unique())
FemalePercent= round((Female2/total_count)*100,2)

#Percentage and Count of Other / Non-Disclosed Players
Other1= purchase_data.groupby(['Gender']).get_group(('Other / Non-Disclosed'))
Other2= len(Other1['SN'].unique())
OtherPercent= round((Other2/total_count)*100,2)

#Make it into a Gender DataFrame
gender1 = {'Total Count':[Male2,Female2,Other2],'Gender':["Male",'Female','Other'],'Percentage of Players':[MalePercent,FemalePercent,OtherPercent]}
gender2 =  pd.DataFrame(gender1)
gender2= gender2.set_index('Gender')
gender2= gender2[['Total Count','Percentage of Players']]
gender2
```

```python
# Group by Gender
grouped_df = purchase_data.groupby(["Gender"])

# Data Manipulation
purchCount = grouped_df["SN"].count()
purchPrice = grouped_df["Price"].mean()
purchValue = grouped_df["Price"].sum()

# Normalize data by deleting all duplicates adn resort
duplicate = purchase_data.drop_duplicates(subset='SN', keep="first")
grouped_dup = duplicate.groupby(["Gender"])

# Find normalized data
purchNorm = (grouped_df["Price"].sum() / grouped_dup["SN"].count())

# Create new DataFrame
Purch_Anal_Gen = pd.DataFrame({"Purchase Count": purchCount,
                              "Average Purchase Price": purchPrice,
                              "Total Purchase Value": purchValue,
                              "Avg Total Purchase per Person": purchNorm})

# DataFrame formatting
Purch_Anal_Gen["Average Purchase Price"] = Purch_Anal_Gen["Average Purchase Price"].map("${:.2f}".format)
Purch_Anal_Gen["Total Purchase Value"] = Purch_Anal_Gen["Total Purchase Value"].map("${:.2f}".format)
Purch_Anal_Gen["Avg Total Purchase per Person"] = Purch_Anal_Gen["Avg Total Purchase per Person"].map("${:.2f}".format)
Purch_Anal_Gen = Purch_Anal_Gen[["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchase per Person"]]
Purch_Anal_Gen
```

```python
bin_age = [0, 9, 14, 19, 24, 29, 34, 39, 1000]

group_names_age = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]



purchase_data["Age Summary"] = pd.cut(purchase_data["Age"], bin_age, labels=group_names_age)

age_df = purchase_data.groupby(["Age Summary"])

count_players_age = age_df["SN"].nunique()
percentage_players_age = age_df["SN"].nunique() / total_count * 100

age_summary_table = pd.DataFrame({"Total Count": count_players_age,
                                  "Percentage of Players": percentage_players_age})

age_summary_table["Percentage of Players"] = age_summary_table["Percentage of Players"].astype(float).map("{:.2f}%".format)

age_summary_table
```

```python
count_purchase_age = age_df["SN"].count()
avg_price_age = age_df["Price"].sum() / count_purchase_age
purchase_value = avg_price_age * count_purchase_age
avg_purchase_person = age_df["Price"].sum() / age_df["SN"].nunique() 

age_purchase_table = pd.DataFrame({"Purchase Count": count_purchase_age,
                                  "Average Purchase Price": avg_price_age,
                                  "Total Purchase Value": purchase_value,
                                  "Avg Total Purchase per Person": avg_purchase_person})

age_purchase_table["Average Purchase Price"] = age_purchase_table["Average Purchase Price"].astype(float).map("${:,.2f}".format)
age_purchase_table["Total Purchase Value"] = age_purchase_table["Total Purchase Value"].astype(float).map("${:,.2f}".format)
age_purchase_table["Avg Total Purchase per Person"] = age_purchase_table["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)

age_purchase_table
```

```python
# Data Manipulation
groupedBySN = purchase_data.groupby(["SN"])
groupedSNCount = groupedBySN["Item ID"].count()
groupedSNTotal = groupedBySN["Price"].sum()
groupedSNAvg = (groupedSNTotal / groupedSNCount)

# Build DF and format
SN_Demo = pd.DataFrame({"Purchase Count": groupedSNCount,
                         "Average Purchase Price": groupedSNAvg,
                         "Total Purchase Value": groupedSNTotal})

SN_Demo = SN_Demo.sort_values("Total Purchase Value", ascending=False) 
SN_Demo["Average Purchase Price"] = SN_Demo["Average Purchase Price"].map("${:.2f}".format)
SN_Demo["Total Purchase Value"] = SN_Demo["Total Purchase Value"].map("${:.2f}".format)
SN_Demo = SN_Demo[["Purchase Count", "Average Purchase Price", "Total Purchase Value"]]
SN_Demo.head()
```

```python
#Most Popular Items



groupItem = purchase_data.groupby(["Item ID", "Item Name"])
groupItemC = groupItem["SN"].count()
groupPriceSum = groupItem["Price"].sum()
groupItemP = (groupPriceSum / groupItemC)
groupItemV = (groupItemP * groupItemC)

# New DF with formatting
Pop_Item = pd.DataFrame({"Purchase Count": groupItemC,
                          "Item Price": groupItemP,
                          "Total Purchase Value": groupItemV})

Pop_Item = Pop_Item.sort_values("Purchase Count", ascending=False) 
Pop_Item["Item Price"] = Pop_Item["Item Price"].map("${:.2f}".format)
Pop_Item["Total Purchase Value"] = Pop_Item["Total Purchase Value"].map("${:.2f}".format)
Pop_Item = Pop_Item[["Purchase Count", "Item Price", "Total Purchase Value"]]
Pop_Item.head()
```
